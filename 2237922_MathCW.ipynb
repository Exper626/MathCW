```python
# imports:
import numpy as np
import matplotlib.pyplot as plt

# The piecewise function:
def piecewise_function(x):
  if x < 0 and x >= -np.pi:
    return (x**2)-1
  elif x >= 0 and x <= np.pi:
    return x*(np.e**(-x/2))
  else:
    return None

# Generate x values
x_values = np.linspace(2*-np.pi, 2*np.pi, 5000)

# Calculate y values for each x
y_values = np.vectorize(piecewise_function)(x_values)

# Plot the graph
plt.plot(x_values, y_values, label='f(x)')
plt.title('Periodic function')
plt.xlabel('x (Domain)')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

### Question 02

Numpy's fast Fourier transform is implemented. The selected function is a simple sinusoidal function. 

a function is defined that generate a sinusoidal signal with a given frequency, sampling rate, and duration.
The sinusoidal signal with a frequency of 10Hz, a sampling rate of 100Hz, and a duration of 1 second is generated.
The graph is plotted

```python
# necessary imports
import numpy as np
import matplotlib.pyplot as plt

# defining the function that generate a sinusoidal signal
def sinusoidal_generation (frequency, sampling_rate, duration):
    t = np.linspace(0, duration, int(duration * sampling_rate))
    signal = np.sin(2 * np.pi * frequency * t)
    return signal

# generate a sinal
frequency = 10
sampling_rate = 100
duration = 1
signal = sinusoidal_generation (frequency, sampling_rate, duration)

# plot the graph
plt.plot(signal)
plt.xlabel('Sample index')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Signal')
plt.grid(True)
plt.show()
```

The DFT is computed using numpy.fft.fft function. The graph is plotted

```python
dft = np.fft.fft(signal)

# plot the graph
plt.plot(np.abs(dft))
plt.xlabel('DFT Coefficient Index')
plt.ylabel('Magnitude')
plt.title('DFT coefficients')
plt.grid(True)
plt.show()
```

The plot shows the magnitude of the DFT coefficient as a function of their indices. Where x-axis represents the coefficient indices and the y-axis represents the magnitude of the coefficients. Aliasing is a phenomenon that occurs when the sampling rate is lower that Nyquist rate, which is twice the highest frequency in the signal. In this example, the signal is a sinusoid with a frequency of 10Hz, and the sampling rate is 100 Hz, which is above the Nyquist rate. Therefore, the signal is sampled without aliasing, and the DFT has a single peak at the 10th coefficient, corresponding to the signals frequency. However, if the sampling rate is reduced to 50 Hz, which is equal to the Nyquist rate, the signal is sampled with aliasing, and the DFT has multiple peaks at the 10th, 50th, 90th, and 130th coefficients, which are the aliases of the signals frequency. This is because the DFT is periodic, and the aliases are the frequencies that are congruent to the signals frequency modulo the sampling rate.

```python
# Define the function
def f(x):
    return np.sin(x)

# Define the variable x
x = np.linspace(-np.pi, np.pi, 400)

# Calculate the Taylor series expansion of f(x) at x = pi/4
taylor_series = np.zeros_like(x)
for n in range(0, 50):
    taylor_series += ((-1)**n / np.math.factorial(2*n + 1)) * ((np.pi/4) - x)**(2*n + 1)

# Plotting the graph
plt.plot(x, f(x), label='sin(x)')
plt.plot(x, taylor_series, label='Taylor series')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor series expansion of sin(x) at x=pi/4')
plt.legend()
plt.grid(True)
plt.show()
```

###### a) logistic function

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
logistics = 1/(1+sp.exp(-x))
logistics
```

```python
# Define the logistic function using Numpy
def logistic_function(x1):
    return 1 / (1 + np.exp(-x1))

# Generate an array of values from -10 to 10 for x-axis
# this is for functions f(x) - logistic and other
x1 = np.linspace(-10, 10, 1000)

# Calculate y values based on the logistic function
y = logistic_function(x1)

# Plotting the graph
plt.plot(x1, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Logistic Function')
plt.grid(True)
plt.show()
print("\n")
```

###### b) derivative of logistic function

```python
# define the variable x using SymPy
xDiff = sp.Symbol('x')

# define the logistic function using SymPy
f = 1 / (1 + sp.exp(-xDiff))

# Calculate the derivative with respect to x using SymPy
f_prime = sp.diff(f, xDiff)

# Convert derivative expression using lambdify method of SymPy
f_prime_func = sp.lambdify(xDiff, f_prime, 'numpy')

# Generate an array of values from -10 to 10 for x-axis
# this is for the derivative f'(x)
x_vals = np.linspace(-10, 10, 1000)

# calculate the y values based on the derivative function
y_vals = f_prime_func(x_vals)

# plotting the graph:
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('f \'(x)')
plt.title('Derivative of Logistic Function')
plt.grid(True)
plt.show()
print("\n")
```
